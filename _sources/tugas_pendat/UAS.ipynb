# UAS

## Membaca Data

```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/ZuniAmandaDewi/dataset/main/milknew.csv")
df
```

```python
df[["pH", "Temprature", "Colour"]].agg(['min','max'])
```

## Preprocessing Data

```python
X = df.drop(columns="Grade")  #data testing
y = df.Grade #data class
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()  #membuat kelas menjadi angka
le.fit(y)
y = le.transform(y)
y
```

```python
le.inverse_transform(y)
```

```python
labels = pd.get_dummies(df.Grade).columns.values.tolist()
labels
```

# Normalisasi Data

```python
from sklearn.preprocessing import MinMaxScaler
judul = X.columns.copy()
scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
#X
hasil = pd.DataFrame(X,columns=judul)
hasil
```

```python
hasil[["pH", "Temprature", "Colour"]].agg(['min','max'])
```

```python
X.shape, y.shape
```

## Split Data

```python
# membagi data menjadi data testing(20%) dan training(80%)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from numpy import array
metode1 = KNeighborsClassifier(n_neighbors=3)
metode1.fit(X_train, y_train)
print(metode1.score(X_train, y_train))
print(metode1.score(X_test, y_test))
y_pred = metode1.predict(array([[6.7, 45, 1, 1, 0, 0, 247]]))
le.inverse_transform(y_pred)[0]
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

metode2 = GaussianNB()
metode2.fit(X_train, y_train)
print(metode2.score(X_train, y_train))
print(metode2.score(X_test, y_test))
y_pred = metode2.predict(array([[6.7, 45, 1, 1, 0, 0, 247]]))
le.inverse_transform(y_pred)[0]
```

## Decesion Tree

```python
from sklearn import tree

metode3 = tree.DecisionTreeClassifier(criterion="gini")
metode3.fit(X_train, y_train)
print(metode3.score(X_train, y_train))
print(metode3.score(X_test, y_test))
y_pred = metode3.predict(array([[6.7, 45, 1, 1, 0, 0, 247]]))
le.inverse_transform(y_pred)[0]
```

## Eksport

```python
from sklearn.utils.validation import joblib
# label encoder
joblib.dump(le, "le.save") 

# scaler
joblib.dump(scaler, "scaler.save") 

# model
joblib.dump(metode1, "knn.joblib")
joblib.dump(metode2, "nb.joblib")
joblib.dump(metode3, "tree.joblib")
```
